```python
import pandas as pd
import os

folder = r"\\cicc.group\DFS\Pub\Workgrp\S_EQ Derivatives\3-部分共享\2-RB Formula\EDSLib_Realtime\EDSLib_Source\EDSLib_Realtime"

# Find the latest file starting with "EDSLib Realtime Result as of"
prefix = "EDSLib Realtime Result as of"
files = [f for f in os.listdir(folder) if f.startswith(prefix) and f.endswith(".xlsx")]
files.sort(reverse=True)
latest_file = files[0]
filepath = os.path.join(folder, latest_file)
print(f"Reading: {latest_file}")

# Read the Excel file - header is on row 3 (0-indexed: row index 2)
df = pd.read_excel(filepath, engine="openpyxl", header=2)

# Filter rows where SubBook contains "QIS"
df_qis = df[df["SubBook"].astype(str).str.contains("QIS", case=False, na=False)].reset_index(drop=True)

print(f"Total rows with QIS in SubBook: {len(df_qis)}")
print(f"Unique SubBook values: {sorted(df_qis['SubBook'].unique())}")
df_qis
```
